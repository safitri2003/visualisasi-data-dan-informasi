```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
data = pd.read_excel('tugas vdi.xlsx')

```

```python
data
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(y='gname', data=data, order=data['gname'].value_counts().iloc[:10].index, palette="coolwarm")
plt.title('Top 10 Most Active Terrorist Groups')
plt.xlabel('Number of Attacks')
plt.ylabel('Group Name')
plt.show()
```

Kutipan kode ini memvisualisasikan 10 kelompok teroris paling aktif dari dataset Anda menggunakan countplot dari library seaborn. Menunjukkan kelompok teroris yang paling sering melakukan serangan di seluruh dunia.
Penjelasan: Kelompok teroris seperti ISIS, Al-Qaeda, dan Taliban sering kali berada di balik banyak serangan. Visualisasi ini menampilkan siapa pelaku utamanya.

```python
# Tren serangan berdasarkan tahun
yearly_trend = data.groupby('iyear').size()

plt.figure(figsize=(10, 6))
plt.plot(yearly_trend.index, yearly_trend.values, marker='o', color='b')
plt.title('Trends in Global Terrorist Attacks by Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

Plot garis : Untuk menunjukkan tren perubahan dari tahun ke tahun.
Area plot : Untuk menunjukkan perubahan dalam jumlah serangan dan memberikan visual yang jelas mengenai peningkatan/penurunan.
